```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("messy_book_list.csv")
df.head()
```

```python
df.columns
```

```python
"""if we want to make a database for books, we should avoid unnecessary details like 'Edition Statement', 'Contributors',
'Corporate Author', 'Corporate Contributors', 'Former owner', 'Engraver', 'Issuance type', 'Shelfmarks', so we will just
drop these values by dropping columns with dataframe.drop() function"""
```

```python
dropped_columns = ['Edition Statement', 'Contributors',
'Corporate Author', 'Corporate Contributors', 'Former owner', 'Engraver', 'Issuance type', 'Shelfmarks']
df.drop(dropped_columns, inplace=True, axis=1)
```

```python
df.head()
```

```python
"""we have to find a unique attribute that users can use to access specific book, with a is_unique function, it is obvious,
that "Identifier" is the one. So, now lets change our index into identifier"""
```

```python
df["Identifier"].is_unique
```

```python
df = df.set_index("Identifier")
df.head()
```

```python
"""lets see the datatype of the rest of the columns, so we can choose the next operation accordingly"""
```

```python
df.dtypes.count
```

```python
"""lets clean non-uniform values in an attribute, lets focus on 'Date of Pun=blication' and 'Place of Publication' for instance"""
```

```python
"""with loc() function we can see the specific values in a column, lets try that for the 'Date of Publication'"""
```

```python
df.loc[1905:, 'Date of Publication'].head(11)
```

```python
"""we will now use regex to clean this column. get a thorough comprehension of this function at 
'https://www.w3schools.com/python/python_regex.asp' before getting dirty with it's operation. find the attached 'meta_char.png'
to see the defined characteristics of metadata used in this function"""
```

```python
"""There has been str.extract used to capture the group of values from regex"""
```

```python
refined = df['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
refined.head()
```

```python
df['Date of Publication'] = pd.to_numeric(refined)
df['Date of Publication'].head()
```

```python
df['Date of Publication'].isnull().sum()/len(df)
```

```python
"""now, lets clean object type 'Place of Publication' column"""
```

```python
df["Place of Publication"]
```

```python
"""as you can see we have common city names nested within unnecessary information, 
for instance, London, Oxford, Newcastle, lets clean these values"""
```

```python
temp = df['Place of Publication']
```

```python
london = temp.str.contains('London')
london.head(10)
```

```python
"""now, lests load the bolean balue in same way for 'Oxford', and, 'Newcastle'"""
```

```python
oxford = temp.str.contains('Oxford')
newcastle = temp.str.contains('Newcastle')
newcastle.tail(10)
```

```python
df['Place of Publication'] = np.where(london, 'London', 
                                       np.where(oxford, 'Oxford', 
                                             np.where(newcastle, 'Newcastle', 
                                                      temp.str.replace('-', ' '))))
df['Place of Publication'].head()
```
